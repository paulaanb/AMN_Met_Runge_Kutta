```python
from metodo_predictor_corrector import euler_predict_correct
import matplotlib.pyplot as plt
```

## Ejemplo
EDO a resolver: $3x^2y' - 4xy = 1$  
Cond. inicial: $y(0.5) = -1$  
Intervalo: $[0.5, 4]$

### Solución numérica

```python
# Definir la ecuación diferencial y'=f(x,y)
def f1(x, y):
    '''EDO a resolver: y'=f1(x,y)'''
    return (1 + 4*x*y)/(3*x**2)

# Parámetros iniciales
x0 = 0.5  # condicion inicial para x
y0 = -1  # condicion inicial para y
xn = 4  # extremo superior del intervalo
h = 0.1  # tamaño del paso

# Llamar al metodo para calular la solución numérica
x_num, y_num = euler_predict_correct(f1, x0, y0, xn, h)

# Representar la solución numérica
plt.plot(x_num, y_num, label='Solución numérica')

plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1) #Eje x
plt.axvline(0, color='black', linewidth=1) #Eje y
plt.title('Solución con Método Predictor-Corrector')
plt.legend()
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

### Error en comparación con la solución analítica
Solución analítica: $$y(x) = -1.8x^{4/3} - \frac{1}{7x}$$

```python
def f1(x, y):
    '''EDO a resolver: y'=f1(x,y)'''
    return (1 + 4*x*y)/(3*x**2)

# Parámetros iniciales
x0 = 0.5  # condicion inicial para x
y0 = -1  # condicion inicial para y
xn = 4  # extremo superior del intervalo
h = 0.1  # tamaño del paso

def sol_analit_1(x):
    '''Solución analítica conocida al problema'''
    return -1.8*x**(4/3) - 1/(7*x)


# Valores solución numérica con el método predictor-corrector
x_num, y_num = euler_predict_correct(f1, x0, y0, xn, h)

# Valores solución analítica (para los mismo valores de x)
y_analit = [sol_analit_1(x) for x in x_num]


# Calcular el error absoluto
error_abs = [abs(y_n - y_a) for y_n, y_a in zip(y_num, y_analit)]
print("Error en el último punto:", error_abs[-1])
```

```python
# Representar la solución numérica
plt.plot(x_num, y_num, label='Solución numérica')

# Representar la solución exacta
plt.plot(x_num, y_analit, label='Solución analítica')

plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1) #Eje x
plt.axvline(0, color='black', linewidth=1) #Eje y
plt.title('Comparación Sol. Analítica vs Sol. Método Predictor-Corrector')
plt.legend()
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

### Comparación con Método Euler básico

Usando el Método de Euler básico el error en el último punto era: $0.6693137794515565$, mucho mayor que usando el Método Predictor-Corrector.
